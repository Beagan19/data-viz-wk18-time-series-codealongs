# Prepare Wide Time Series

```python
# !pip install yfinance
import yfinance as yfin
yfin.pdr_override()

##  Pandas DataReader Import
import pandas_datareader.data as pdr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
import datetime as dt
start_date = '2001-01-01'
end_date = dt.date.today()
end_date
```

```python
cryptocurrencies = {"Bitcoin":'BTC-USD',#"Tether":'USDT-USD',
                    "Etherium":'ETH-USD',"Dogecoin": "DOGE-USD"}
```

```python
crypto_dfs = []
crypto_df_dict = {}
## Loop through dict of currency names and symbols
for currency, symbol in cryptocurrencies.items():
    
    temp_df = pdr.get_data_yahoo(symbol, start_date, end_date)[['Adj Close']]
    
    temp = temp_df['Adj Close']
    temp.name = currency
    crypto_dfs.append(temp)
    
    temp_df['Currency']=currency
    temp_df = temp_df.rename({'Adj Close':symbol},axis=1)
#     temp_df['Symbol'] = symbol
    crypto_df_dict[currency] = temp_df

df = pd.concat(crypto_dfs, axis=1)
df
```

```python
df_wide = df.T
df_wide.columns = df_wide.columns.strftime('%Y-%m-%d')

df_wide = df_wide.reset_index()
df_wide = df_wide.rename({'index':'CryptoCurrency'},axis=1)
df_wide
```

```python

df_wide['Symbol'] = df_wide['CryptoCurrency'].map(cryptocurrencies)
df_wide['Currency'] = df_wide['Symbol'].map(lambda x: x.split('-')[0])
df_wide
```

```python
id_cols = ['CryptoCurrency','Symbol','Currency']
df_wide = df_wide[[*id_cols,*df_wide.drop(columns=id_cols).columns]].copy()
df_wide
```

```python
# df_wide = df_wide.rename({'index':'Symbol'},axis=1)
df_wide = df_wide.reset_index(drop=True)
df_wide
```
